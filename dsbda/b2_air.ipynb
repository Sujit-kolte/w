# B2 Air Quality

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
df = pd.read_csv('b2_air_dataset.csv')
df.head()
```

## Data Cleaning

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
print("Duplicates Removed")
```

## Data Integration

```python
df2 = df.copy()
merged = pd.concat([df,df2])
merged.head()
```

## Data Transformation

```python
df['AQI'] = df['AQI'] * 1.1
df.head()
```

## Error Correcting

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

## Data Model Building

```python
x = df[['PM2.5','PM10','NO2']]
y = df['AQI']

model = LinearRegression()
model.fit(x,y)

print("Model Built Successfully")
```
