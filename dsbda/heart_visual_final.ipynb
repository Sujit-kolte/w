# Heart Disease Visualization

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```
df=pd.read_csv('heart_dataset_final.csv')
df.head()
```

```
df.shape
```

```
df.info()
```

```
df.isna().sum()
```

```
df['Sex'].replace(['M','F'],[1,0], inplace=True)
df['ChestPainType'].replace(['ATA','NAP','ASY','TA'],[0,1,2,3], inplace=True)
df.head()
```

```
sns.countplot(x='HeartDisease',data=df)
plt.show()
```

```
sns.countplot(x='Sex',data=df)
plt.show()
```

```
sns.displot(df['Age'],kde=True)
plt.show()
```

```
sns.scatterplot(x='Age',y='Cholesterol',data=df)
plt.show()
```

```
group=df.groupby('Sex')['Cholesterol'].mean().reset_index()
sns.barplot(x='Sex',y='Cholesterol',data=group)
plt.show()
```

```
plt.hist(df['MaxHR'])
plt.show()
```

```
sns.boxplot(y='RestingBP',data=df)
plt.show()
```

```
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='Reds')
plt.show()
```
