# Air Quality Visualization

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```
df=pd.read_csv('airquality_dataset_final.csv')
df.head()
```

```
df.shape
```

```
df.info()
```

```
df.isna().sum()
```

```
sns.countplot(x='City',data=df)
plt.show()
```

```
sns.displot(df['AQI'],kde=True)
plt.show()
```

```
sns.scatterplot(x='CO_GT',y='AQI',data=df)
plt.show()
```

```
group=df.groupby('City')['AQI'].mean().reset_index()
sns.barplot(x='City',y='AQI',data=group)
plt.show()
```

```
plt.hist(df['NO2_GT'])
plt.show()
```

```
sns.boxplot(y='Temperature',data=df)
plt.show()
```

```
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='Greens')
plt.show()
```
